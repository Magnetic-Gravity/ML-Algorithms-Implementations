# Gradient Descent Taking Slope(M) Constant
- Gradient descent is an iterative optimization algorithm, crucial in machine learning, that finds the minimum of a function (like a cost/loss function) by taking repeated steps in the opposite direction of the function's gradient (steepest descent), adjusting model parameters (weights/biases) to minimize errors between predictions and actual data, effectively training the model to improve accuracy.
- It's like walking downhill in the fog by feeling the slope to find the valley floor, adjusting steps based on the gradient's direction and magnitude

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Lets apply OLS(Ordinary Least Squares)
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='red')
```

```python
# Lets apply Gradient Descent assuming Slope/Weight/Coefficient is constant (m = 78.35) and
# Lets assume the starting value for Intercept/bias b=0 # Any value initialization of b will give same result 

y_pred = ((78.35 * X) + 0).reshape(4)     # Reshape(4) Makes it a 1-D array with 4 elements
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='#00a65a', label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 1 
m = 78.35
b = 0
loss_slope = -2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
# Lets take learning rate = 0.1
lr = 0.1
step_size = loss_slope * lr
step_size
```

```python
#calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred1, color='#00a65a', label='b = {}'.format(b))
plt.plot(X, y_pred, color='#A3E4D7', label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration 2 

loss_slope = -2*np.sum(y-m*X.ravel()-b)
print('Loss Slope :',loss_slope) 

step_size = loss_slope * lr
print('Step Size : ', step_size)

b = b - step_size
print('bias b :', b)


y_pred2 = ((78.35 * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred2, color='#A3E4D7', label='b = {}'.format(b))
plt.plot(X, y_pred1, color='#00a65a', label='b = {}'.format(b))
plt.plot(X, y_pred, color='#A3E4D7', label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration 3

loss_slope = -2*np.sum(y-m*X.ravel()-b)
print('Loss Slope :',loss_slope) 

step_size = loss_slope * lr
print('Step Size : ', step_size)

b = b - step_size
print('bias b :', b)


y_pred3 = ((78.35 * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred3, color='#00a65a', label='b = {}'.format(b))
plt.plot(X, y_pred2, color='#A3E4D7', label='b = {}'.format(b))
plt.plot(X, y_pred1, color='#A3E4D7', label='b = {}'.format(b))
plt.plot(X, y_pred, color='#A3E4D7', label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration 4

loss_slope = -2*np.sum(y-m*X.ravel()-b)
print('Loss Slope :',loss_slope) 

step_size = loss_slope * lr
print('Step Size : ', step_size)

b = b - step_size
print('bias b :', b)


y_pred4 = ((78.35 * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred4, color='#00a65a', label='b = {}'.format(b))
plt.plot(X, y_pred3, color='#A3E4D7', label='b = {}'.format(b))
plt.plot(X, y_pred2, color='#A3E4D7', label='b = {}'.format(b))
plt.plot(X, y_pred1, color='#A3E4D7', label='b = {}'.format(b))
plt.plot(X, y_pred, color='#A3E4D7', label='b=0')
plt.legend()
plt.show()
```

```python
#Normal Looping of this
b = -100
m = 78.35
lr = 0.1

epochs = 10
for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*X.ravel() - b)
    b = b - (lr * loss_slope)    
    y_pred = m * X + b
    plt.plot(X,y_pred)
    
plt.scatter(X,y)
```

```python
#By taking vary small Learning Rate 
b = -100
m = 78.35
lr = 0.01 

epochs = 10
for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*X.ravel() - b)
    b = b - (lr * loss_slope)    
    y_pred = m * X + b
    plt.plot(X,y_pred)
    
plt.scatter(X,y)
```

# Making a Class Named as GD Regressor

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1, noise=20)
```

```python
plt.scatter(X,y)
```

```python
# Lets apply OLS(Ordinary Least Squares)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
# Slope
m = 27.82
```

```python
class GDRegressor:

    def __init__(self, learning_rate, epochs):
        self.m = 27.82
        self.b = -120
        self.lr = learning_rate
        self.epochs = epochs 

    def fit(self,X,y):
        #calculate 'b' using Gradient Descent(GD)
        for i in range(self.epochs):
            loss_slope = -2 * np.sum(y - m*X.ravel() - self.b)
            self.b = self.b - (self.lr * loss_slope)
            print(loss_slope, self.b)   
        print(self.b)
```

```python
gd = GDRegressor(0.001,100)
gd.fit(X,y)
```
