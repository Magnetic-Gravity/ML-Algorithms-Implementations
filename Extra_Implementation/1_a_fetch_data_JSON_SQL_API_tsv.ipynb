### Run A For Loop and Store all API_DATA into a data-frame named df1 and then create a CSV file from this data

```python
import pandas as pd
import requests

df1 = pd.DataFrame()

for i in range(1, 12):
    url = f"https://newsdata.io/api/1/latest?apikey=pub_42cb9b589e374144b529b3b55fc4bc6d&q={i}"
    response = requests.get(url)
    results = response.json().get('results', [])
    if isinstance(results, dict):
        results = [results]
    temp = pd.DataFrame(results)
    df1 = pd.concat([df1, temp], ignore_index=True)
```
