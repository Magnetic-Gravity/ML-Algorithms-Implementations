# Univariate analysis
Univariate analysis is the simplest form of data analysis, focusing on one variable at a time to describe 
its characteristics, find patterns, and understand its distribution, using tools like means, medians, modes 
(central tendency), ranges, standard deviations (dispersion), and visualizations such as histograms or box plots, 
without examining relationships between variables.

### Data Types (1) Numerical Data ex:height,age  (2) Categorical Data ex:nationality,gender

```python
import pandas as pd
df = pd.read_csv("train.csv")
df.head()
```

# 1. Categorical Data

### a. Countplot
This function in the Seaborn library is used to display the count of observations for each category within a categorical variable

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='Survived', data=df)
plt.title('Count of Passengers Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

df['Survived'].value_counts()
```

```python
# They both work in the same way, the upper one and this one 
# df['Survived'].value_counts().plot(kind='bar')
# df['Survived'].value_counts()

df['Pclass'].value_counts().plot(kind='bar', color=['red', 'blue', 'green'])
df['Pclass'].value_counts()
```

```python
df['Embarked'].value_counts().plot(kind='bar', color=['red', 'blue', 'green'])
df['Embarked'].value_counts()
```

## b. PieChart
A pie chart is a circular graph divided into slices (sectors) that visually represent numerical proportions or percentages of a whole, where each slice's size is proportional to the quantity it represents, making it easy to see how parts contribute to a total. All slices combine to form a complete circle (100% or 360), ideal for showing composition and parts-to-whole relationships in simple datasets. 

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f')
df['Survived'].value_counts()
```

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%.2f')
df['Pclass'].value_counts()
```

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%.2f')
df['Sex'].value_counts()
```

# Numerical Data

## a. Histogram

A histogram is a graphical tool that displays the frequency distribution of continuous numerical data by grouping it into "bins" or intervals, shown as adjacent bars where the height indicates frequency, helping visualize patterns, trends, and spread in large datasets. Unlike bar charts for categories, histograms use a continuous X-axis for data ranges, revealing the shape of data distribution (like normal, skewed) and identifying outliers.

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=30)            #bins = no. building type shape
```

## b. Distplot

The distplot (distribution plot) is a function from the Python Seaborn library used to visualize the univariate distribution of observations for continuous data. It combines a histogram with a Kernel Density Estimate (KDE) curve (and optionally a rug plot) to show both the frequency of data points within bins and a smoothed estimate of the underlying probability density function. 

```python
import seaborn as sns
sns.distplot(df['Age'])
```

## c. Boxplot

A box plot (or box-and-whisker plot) visually summarizes a dataset's distribution using five key numbers: minimum, first quartile (Q1), median, third quartile (Q3), and maximum, showing spread and identifying potential outliers. The central box represents the middle 50% of data (the Interquartile Range, IQR), with a line for the median, while whiskers extend to show variability outside the box, and individual points mark outliers.  

```python
import seaborn as sns
sns.boxplot(df['Fare'])
```

```python
sns.boxplot(df['Age'])
```

```python
print(df['Age'].min())
print(df['Age'].max())
print(df['Age'].mean())
```

```python
# 0: Perfectly Symmetrical,  # +ve: positively skewed right or up, -ve:negatively skewed towards left or down 
print(df['Age'].skew())  
```
